## Preparing Labels for Visualisation 

To visualise the separation of our labels, we need labels! For this project, there will be a single binary label, based on  the outcome rating that we are interested in: _"How often home health patients had to be admitted to the hospital"_. 

This outcome rating is numeric and normally-distributed, so any agency whose rating lay more than two standard deviations from the mean (on the wrong side!) was defined as a "poorest performing" agency for that outcome. These agencies received the label '1' and all other agencies received the label '0'.  The mean rating for our outcome was 15.217 and the standard deviation was 4.224, so the label threshold was 23.665 (note: in this case, higher frequency of hospitalisation denotes poorer performance). The statistical calculation & label creation was done in Microsoft Excel and merged into the dataset.
 
In our dataset, 274 agencies came out as the "poorest performing" for this outcome and received the label '1'.

```python
# Use a .py file to keep the import statements and data loading out of the narrative
%run exploratory_data_analysis_setup.py
print 'There are %1i samples in the dataset (having eliminated samples where the target label was null).' % (dataset.shape[0])
print 'There are %1s poorly performing agencies identified in the dataset by the label \'%2s\'.' % ((sum(p == 1 for p in labels)), target_label)
```

```python
import pandas
dataset = pandas.read_csv("./dataset_v5.csv")

# Select the target lable and strip out any rows where there is no label to train or test with
target_label = 'Lbl: Poor Hosp Rating 1SD'

dataset = dataset[dataset[target_label].notnull()]

samples = dataset[[ # keep the agency behavior fields we're interested in
    'How often the home health team began their patients care in a timely manner',
    'How often the home health team taught patients (or their family caregivers) about their drugs',
    'How often the home health team checked patients risk of falling',
    'How often the home health team checked patients for depression',
    'How often the home health team made sure that their patients have received a flu shot for the current flu season.',
    'How often the home health team made sure that their patients have received a pneumococcal vaccine (pneumonia shot).',
    'With diabetes - how often the home health team got doctors orders and gave foot care and taught patients about foot care',
    'How often the home health team checked patients for pain',
    'How often the home health team treated their patients pain',
    'How often the home health team took doctor-ordered action to prevent pressure sores (bed sores)',
    'How often the home health team checked patients for the risk of developing pressure sores (bed sores)']]

# (1) reshape the label matrix to be flat and (2) change the boolean {false, true} values into {0,1} as sklearn modules expect
import numpy as np
from sklearn import preprocessing
labels = np.reshape(dataset[target_label].astype(int), -1)
samples = preprocessing.normalize(samples.astype(float))
samples = preprocessing.scale(samples)
```

### A brief (but important) note on label imbalance   

The extreme imbalance in the label frequencies highlights that 'accuracy' would not be a useful measure of performance for our classifier. Consider a trivial classifier that predicts _"no agencies perform poorly"_ - it would have an accuracy of 97%!  In contrast, such a classifier would also produce precision & recall scores of zero - because it would fail to identify *any* poorly performing agencies.

## Visualising The Data

A pair-wise set of scatter plots (coloured by label) was generated using [Weka](http://www.cs.waikato.ac.nz/ml/weka/) _(a workbench for data analysis and Machine Learning, developed by the University of Waikato, New Zealand)._ These were analysed to see if either of the following could be identified:
* Correlations between features
* Separability in the target label distribution

A representative sample of these scatter plots is shown below, using python & matplotlib.

###### Some admin... Plotting functions for use throughout this notepad

```python
%matplotlib inline
from matplotlib import pyplot
data_labels, markers, colours, alphas = [0, 1], ['o', 'o'], ['b', 'r'], [0.3, 0.6]
def scatterplot(xaxis, yaxis, y, title = '', xlabel = 'X', ylabel = 'Y'):
    for label, marker, colour, a in zip(data_labels, markers, colours, alphas):
        pyplot.scatter(xaxis[y == label], yaxis[y == label], c=colour, s = 30, cmap=pyplot.cm.bwr, marker=marker, alpha=a, edgecolors='none')
    pyplot.xlabel(xlabel)
    pyplot.ylabel(ylabel)
    pyplot.title(title)
    pyplot.legend(loc='best')
    pyplot.show()
```

## Correlations Between Features

There is only one apparent correlation between features in the set - a relationship between two types of shots/vaccines (flu shot and pneumonia shot). This is not particularly surprising or interesting, but it is the only apparent pair-wise correlation in the feature set.

### Principal Component Analysis

As the features offered no obvious 2-dimensional separation of the labels, it was necessary to look to other statistical techniques to help and first among these was Principal Component Analysis (PCA). PCA is a commonly-used technique that helps identify patterns in a dataset by projecting the features on to a set of lower-dimensional component axes - those that maximise the variation in the dataset.

**Data Wrangling:** 
As simple as the PCA is in theory, in practise it required a new round of data wrangling.  The scikit-learn libraries would not accept null values in any of the features used in the analysis. Values then had to be imputed for all samples that contained null values.  Each null value was replace with the mean value for that feature, as this seemed to be a reasonably well-respected strategy for imputing values.  Depending on the requirements, a K-Nearest Neighbour imputation strategy also [appears to be well-respected](http://scikit-learn.org/stable/modules/preprocessing.html#imputation-of-missing-values). However, as can be seen from the scatter plots above, it was not clear that a nearest-neighbour imputation would be valid for this data.

**Visualising the PCA:** 
The first two principal components were plotted against each other with the data points coloured to match the target label. The hope was that this might show some visible separation of the labels. Unfortunately, that did not turn out to be the case, as can be seen from the diagram below:

```python
from sklearn.decomposition import PCA
components = PCA(n_components=2).fit_transform(samples)
scatterplot(components[:,0], components[:,1], labels, 'Principal Component Analysis (PCA)', '1st Principal Comp.', '2nd Principal Comp.')
```

### Linear Discriminant Analysis

Linear Discriminant Analysis (LDA) is another statistical technique that can help to separate labels. Where PCA maximises the variance in the overall dataset, LDA looks at the variances with respect to the particular label we are interested in. 

**Issues with LDA:** I encountered some problems when doing the LDA analysis, chiefly that I was unable to get sklearn.LDA to reduce the 11 dimensions down to anything higher than 1 dimension. I believe it should be possible to do so, but I was unable to achieve it (despite much searching).

**Visualising the LDA:** The single Linear Discriminant was plotted on a line, with the data points again coloured to match the target label.  Unfortunately, as before, it produced no obvious separation between the classes (as can be seen from the diagram below).

```python
from sklearn.lda import LDA
components = LDA(n_components=2).fit_transform(samples, labels)
# LDA has reduced the dimensions to 1, so we have to plot on a line
scatterplot(components[:,0], np.zeros(labels.shape), labels, 'Linear Discriminant Analysis (LDA)', 'Component 1', 'There is no Component 2')
```

### _Footnote: LDA is not applicable to this data_

Later in the project, I had cause to revisit the Linear Discriminant Analysis technique and realised that it is predicated on the independent variables being normally distributed.  I was fairly sure that they were not, and indeed a QQ plot of each feature confirmed this (an example of one of the features is provided below). So the LDA technique would not have been suitable in any case.

```python
import scipy.stats as stats
from matplotlib import pylab
stats.probplot(dataset['How often the home health team checked patients for depression'], dist="norm", plot=pylab)
pylab.show()
```
